```python
import matplotlib.pyplot as plt
def plot_survival_bars(performance, title):
    
    # Create bars & The X position of bars
    objects = ('No', 'Yes')
    y_pos = np.arange(len(objects))

    # Create barplot
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)

    plt.title(title)
    plt.xlabel('Survived ?')
    plt.ylabel('Count')

    # Create labels

    # Text on the top of each barplot
    for i in range(len(y_pos)):
        plt.text(x = y_pos[i]-0.05 , y = performance[i]+0.5, s = performance[i], size = 10)

    # Adjust the margins
    plt.subplots_adjust(bottom= 0.2, top = 0.98)

    # Show graphic
    plt.show()

```

```python
import re
def find_maiden_name(x):
    try:
        return re.search('.*(\()(.*)(\s)(.*)(\))', x, re.IGNORECASE).group(4)
    except AttributeError:
        return ''
    

surname_name = all.Name[all.Title=='Miss'].apply(lambda x: x.split(',')[0].strip())
maiden_name  = all.Name[all.Title=='Mrs'].apply(lambda x: find_maiden_name (x))        

 # checking that we don't have a surname and maiden name for the same person
test = all.assign(Maiden_name=maiden_name, Surname_name=surname_name)
test[test.Maiden_name.notna() & test.Surname_name.isna()]

# Adding the new column : combination of surname and maiden_name
all = all.assign(MaidenSurname=maiden_name.append(surname_name))
```

# Modelling

### Helper functions to compute metrics and save predictions

Recap : 

$$ Accuracy = \frac{TP + TN}{ALL Predictions}$$
The true ones taking into consideration only our prediction
$$ Precision = \frac{TP}{ALL\_Predicted\_Pos} $$ 
The true ones taking into consideration what exists in reality
$$ Sensitivity  = true positive rate = Recall = \frac{TP}{ALL\_Existing\_Pos} $$

$$ F1\_score = \frac{2PR}{P+R} $$

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score, precision_score, recall_score

def get_metrics(train_y, predictions):
    # Model Accuracy: how often is the classifier correct?
    a, p, r = accuracy_score(train_y, predictions), precision_score(train_y, predictions), recall_score(train_y, predictions)
    print("Accuracy:", a)
    print("Precision :", p)
    print("Recall :", r)
    return a, p,r
```

### ROC curve
Receiver Operating Characteristic(ROC) curve is a plot of the true positive rate against the false positive rate. It shows the tradeoff between sensitivity and specificity.

* axis=1 (columns/ y_axis) : True Positive Rate = Recall

* axis=0 (rows/ x_axis)  : $$ False Positive Rate = \frac{FP}{AllNeg} $$

Dans la théorie de la détection du signal, les grandeurs d' et A' mesurent l'aire sous la courbe ROC. Plus cette aire est grande, plus la courbe s’écarte de la ligne du classificateur aléatoire et se rapproche du coude du classificateur idéal (qui passe de (0, 0) à (0, 1) à (1, 1))

```python
from sklearn.metrics import roc_curve, roc_auc_score
def plot_roc(y_test, y_pred_proba):
    fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
    auc = roc_auc_score(y_test, y_pred_proba)
    plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
    plt.legend(loc=4)
    plt.show()
```

```python
def save(predictions, method):
    predictions = list(map(int, predictions)) 
    predictions = pd.DataFrame(predictions, columns=['Survived'])
    test_passenger_id = pd.read_csv('../input/test.csv').iloc[:, 0]
    predictions = pd.concat((test_passenger_id, predictions), axis = 1)
    predictions.to_csv('y_test_{}.csv'.format(method), sep=",", index = False)
```
